### About this Notebook

This notebook reviews how to read and load data from plain text files.

---

## 1. Reading plain text files

Using [numpy](https://numpy.org), it is possible to read from a plain text file and load it directly into an array.

```python
import numpy as np

data = np.loadtxt('example01.txt', comments='#')
data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.plot(data[:,0],data[:,1], color='crimson')
plt.show()
```

When the data in the file is separated by a symbol other than a space, it must be written with the `delimiter` argument.

```python
import numpy as np

data = np.loadtxt('example02.txt', comments='#')
```

```python
import numpy as np

data = np.loadtxt('example02.txt', comments='#', delimiter=',')
data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.plot(data[:,0],data[:,1], color='crimson')
plt.show()
```

## 02. Writing plain text files

Again, the [numpy](https://numpy.org) library has a simple function for writing plain text files.

```python
def f(x):
    return np.exp(-x**2/5)

wdata = np.zeros([2,100])

wdata[0,:] = np.linspace(-10,10,100)
wdata[1] = f(wdata[0])


np.savetxt('out_file1.txt', wdata, fmt='%5.5f', header='This is the header of the file')
```

```python
outdata = np.loadtxt('out_file1.txt', comments='#')


plt.figure(figsize=(7,5))
plt.plot(outdata[0],outdata[1], color='crimson')
plt.show()
```

```python
np.savetxt('out_file2.data', wdata, delimiter = ',', 
           fmt='%5.5f', header='This is the header of the file')

outdata2= np.loadtxt('out_file2.data', comments='#', delimiter=',')


plt.figure(figsize=(7,5))
plt.plot(outdata2[0],outdata2[1], color='crimson')
plt.show()
```

### Open .dat+ReadMe files.

To access the information in the flat file .dat file (in ascii format) and the metadata in the ReadMe file, the command [astropy.io.ascii](https://docs.astropy.org/en/stable/io /ascii/)

```python
import numpy as np
from astropy.io import ascii
import matplotlib.pyplot as plt


data = ascii.read('table1.dat', readme='ReadMe')
data
```

The length of the table is 88 samples and 12 features, including 'Name', 'z', 'sigma*', etc.

---
It is possible to access each of the features,

```python
data['sigma*']
```

```python
data['logL']
```

Full data for a particular sample or set of samples can also be accessed,

```python
data[5]
```

```python
data[[0,1,3,4]]
```

We will distribute the information in a set of numpy arrays,

```python
z = np.array(data["z"])
sigma_star = np.array(data["sigma*"])
e_sigma_star = np.array(data["e_sigma*"])
logL = np.array(data["logL"])
e_logL = np.array(data["e_logL"])
logM = np.array(data["logM"])
e_logM = np.array(data["e_logM"])
```

## 2. View information

Let's look at different combinations of features, looking for some kind of pattern of interest,

```python
plt.scatter(z, logM,  color='crimson')
plt.xlabel(r'$z$')
plt.ylabel(r'$\log M$')
plt.show()
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(221)
plt.plot(z, logM, 'b.')
plt.xlabel(r'$z$')
plt.ylabel(r'$\log M$')

plt.subplot(222)
plt.plot(logL, logM, 'r.')
plt.xlabel(r'$\log L$')
plt.ylabel(r'$\log M$')

plt.subplot(223)
plt.plot(z, logL, 'g.')
plt.xlabel(r'$z$')
plt.ylabel(r'$\log L$')

plt.subplot(224)
plt.plot(sigma_star, logM, 'k.')
plt.xlabel(r'$\sigma_{*}$')
plt.ylabel(r'$\log M$')

plt.show()
```

Some interesting information or patterns appear. For example,

- Two separate points (outcaast) are observed in the graph of $\log M$ vs $z$.
- It seems that there are two separate clusters on the graph of $\log M$ vs $\log L$
- It looks like there are three separate clusters (or two separate clusters and two separate points) on the graph of $\log L$ vs $z$
- There is a correlation trend in the graph of $\log M$ vs $\sigma_{*}$

Although these behaviors seem to exist, some of them should be carefully reviewed. For example, the apparent clusters in figures 2 and 3 correspond to a lack of data, misinterpreted when converting the information to arrays. Reviewing the information in $\log L$ we have

```python
logL
```

where a large number of zeros values are observed that actually corresponded to missing data in the initial set. These zero values were added using the `logL = np.array(data["logL"])` command.

NOTE: It is important to properly handle missing data!

---
On the other hand, we can analyze in more detail the apparent correlation between $\log M$ and $\sigma_{*}$

```python
plt.plot(sigma_star, logM,  'k.')
plt.xlabel(r'$\sigma_{*}$')
plt.ylabel(r'$\log M$')
plt.show()
```

It may look like a linear dependency between the two variables. However, to have a better appreciation, we will include the error bars. For this we use the function [matplotlib.pyplot.errorbar](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.errorbar.html)

```python
plt.errorbar(sigma_star, logM, e_logM, e_sigma_star, fmt='k.')
plt.xlabel(r'$\sigma_{*}$')
plt.ylabel(r'$\log M$')
plt.show()
```

This visualization shows a non-linear correlation. To verify this, we can modify the graph including the logarithm of the velocity dispersion, $\log M$ vs $\log \sigma_{*}$, with which the linear trend is evident,

```python
plt.plot(np.log10(sigma_star), logM,  'k.')
plt.xlabel(r'$\log \sigma_{*}$')
plt.ylabel(r'$\log M$')
plt.show()
```

To confirm this appreciation, we again include the associated uncertainties. For this it is necessary to remember the propagation of errors,

\begin{align}
f(x) =& \log_{10} x = \frac{\ln x}{\log_{10}}\\
\Delta f = &\frac{1}{\log_{10}} \frac{\Delta x}{x}
\end{align}

```python

e_log_sigma = e_sigma_star/(np.log(10.)*sigma_star)


plt.errorbar(np.log10(sigma_star), logM, e_logM, e_log_sigma, fmt='k.')
plt.xlabel(r'$\log \sigma_{*}$')
plt.ylabel(r'$\log M$')
plt.show()
```
